Category-level Analysis

```python
# Total sales by category
print(df.groupby('Category')['Sales'].sum())

# Total profit by category
print(df.groupby('Category')['Profit'].sum())
```

Visualization (Seaborn/Matplotlib)

```python
# Sales per Category
sns.barplot(x='Category', y='Sales', data=df, estimator=sum)
plt.title("Total Sales by Category")
plt.show()

# Profit per Category
sns.barplot(x='Category', y='Profit', data=df, estimator=sum)
plt.title("Total Profit by Category")
plt.show()
```

Discount vs Profit

```python
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title("Discount vs Profit")
plt.show()
```

Compute Correlation Matrix

```python
corr = df[['Sales','Profit','Discount','Quantity']].corr()
print(corr)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap")
plt.show()
```

Time-Series Analysis (Trends Over Time)

```python

df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month

monthly_sales = df.groupby(['Year','Month'])['Sales'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(x='Month', y='Sales', hue='Year', data=monthly_sales, marker='o')
plt.title("Monthly Sales Trend Over Years")
plt.show()
```

Regional & Segment Analysis

```python
# Sales by Region
sns.barplot(x='Region', y='Sales', data=df, estimator=sum)
plt.title("Total Sales by Region")
plt.show()

# Profit by Segment
sns.barplot(x='Segment', y='Profit', data=df, estimator=sum)
plt.title("Total Profit by Customer Segment")
plt.show()
```

Category & Sub-Category Deep Dive

```python
# Sales by Sub-Category
plt.figure(figsize=(12,6))
sns.barplot(x='Sub-Category', y='Sales', data=df, estimator=sum)
plt.title("Sales by Sub-Category")
plt.xticks(rotation=45)
plt.show()

# Profit by Sub-Category
plt.figure(figsize=(12,6))
sns.barplot(x='Sub-Category', y='Profit', data=df, estimator=sum)
plt.title("Profit by Sub-Category")
plt.xticks(rotation=45)
plt.show()
```

Discount vs Profit Analysis

```python
sns.scatterplot(x='Discount', y='Profit', data=df, alpha=0.5)
plt.title("Discount vs Profit")
plt.show()
```

Top Customers & Products

```python
# Top 10 customers by Sales
top_customers = df.groupby('Customer Name')['Sales'].sum().nlargest(10)
print(top_customers)

# Top 10 products by Profit
top_products = df.groupby('Product Name')['Profit'].sum().nlargest(10)
print(top_products)
```
